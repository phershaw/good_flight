# Exploratory Data Analysis

There are two main observations from this analysis: 

1. Reviews tend to be polarized, and slightly skewed negative. 
2. People who buy fancy pants tickets enjoy their flight more.

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()  # Register converters for datetime objects
```

```python
#Load cleaned data
df = pd.read_csv('/Users/paulhershaw/brainstation_course/test_git/test/data/airline_reviews_cleaned.csv')
df
```

```python
#Describe the data
df.describe(include='all').transpose()
```

```python
# Selecting only numeric columns for histograms
numeric_cols = df.select_dtypes(include=['number']).columns

# Creat a figure with subplots for each numeric column
num_plots = len(numeric_cols)
fig, axes = plt.subplots(num_plots, 1, figsize=(10, num_plots * 4))

# Plotting a histogram for each numeric column
for i, col in enumerate(numeric_cols):
    sns.histplot(df[col], ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Adjusting the layout for better readability
plt.tight_layout()
plt.show()
```

## Observation about Histograms

You can see from the reviews that sentiment skews negative, and is centred around the poles as opposed to the mean. This is normal for reviews, but does present a challenge, as it signals that many people have normal flights, are not motivated to leave a review. You can also see this trend below, which breaks down overall rating by recommendation. 

```python
# Plot stacked bar chart
stacked_data = df.groupby(['overall', 'recommended']).size().unstack()
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Overall Rating Broken Down by Recommendation')
plt.xlabel('Overall Rating')
plt.ylabel('Number of Reviews')
plt.show()
```

```python

# List of columns to create horizontal bar charts for
columns = ['traveller_type', 'cabin', 'verified_review']

# Creating a horizontal bar chart for each column
fig, axes = plt.subplots(nrows=len(columns), ncols=1, figsize=(10, 10))
for i, column in enumerate(columns):
    # Grouping by the column and 'sentiment', then calculating the ratio
    sentiment_ratio = df.groupby(column)['sentiment'].value_counts(normalize=True).unstack().fillna(0)

    # Find the sentiment with the highest ratio for each category and sort by it
    dominant_sentiment = sentiment_ratio.idxmax(axis=1)
    sorted_indices = sentiment_ratio.apply(lambda row: row[dominant_sentiment[row.name]], axis=1).sort_values(ascending=False).index
    sentiment_ratio_sorted = sentiment_ratio.loc[sorted_indices]

    # Plot bar chart
    sentiment_ratio_sorted.plot(kind='barh', stacked=True, ax=axes[i])
    axes[i].set_title(f'Sentiment Ratio by {column.capitalize()}')
    axes[i].set_xlabel('Ratio')
    axes[i].set_ylabel(column.capitalize())

plt.tight_layout()
plt.show()
```

## Sentiment Ratio
Notably, first class and business class travellers are more likely to enjoy their flights. 

```python
# List of columns to create horizontal bar charts for
columns = ['traveller_type', 'cabin', 'verified_review', 'airline', 'home_country']

# Set up the figure for multiple plots
fig, axes = plt.subplots(nrows=len(columns), ncols=1, figsize=(10, 40))

# Create a horizontal bar chart for each column
for i, column in enumerate(columns):
    # Grouping by the column and 'sentiment', then calculating the count
    sentiment_count = df.groupby(column)['sentiment'].value_counts().unstack().fillna(0)

    # Sorting the categories based on the total number of reviews
    sentiment_count['total_reviews'] = sentiment_count.sum(axis=1)
    sentiment_count_sorted = sentiment_count.sort_values(by='total_reviews', ascending=False)
    
    # Dropping the total count column for plotting
    sentiment_count_sorted_plot = sentiment_count_sorted.drop('total_reviews', axis=1)

    # Plotting the bar chart
    sentiment_count_sorted_plot.plot(kind='barh', stacked=True, ax=axes[i])
    axes[i].set_title(f'Total Number of Reviews by Sentiment for {column.capitalize()}')
    axes[i].set_xlabel('Number of Reviews')
    axes[i].set_ylabel(column.capitalize())

plt.tight_layout()
plt.show()
```
